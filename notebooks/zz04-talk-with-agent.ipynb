```python
from letta_client import Letta

client = Letta(base_url="http://localhost:8283")
```

```python
agent_id = "agent-d92f2e63-7a62-4b34-8ed8-fca3a5bbe20c"
agent = client.agents.retrieve(
    agent_id=agent_id,
)
```

```python
from utils.agent_info import pretty_print_messages

response = client.agents.messages.create(
    agent_id=agent.id,
    input="你好，我是胡德扬"
)

# Use our new pretty parser instead of the messy single print!
pretty_print_messages(response.messages)
```

```python
# If we want to see the whole historical log properly:
# Get all messages
history = list(client.agents.messages.list(agent_id=agent_id))

# the Letta return from list() usually is naturally reversed (oldest first or newest first). Let's print the last 5
pretty_print_messages(history[-5:])
```

```python
# Let's prompt the agent to explicitly use a tool (like update core memory) so we can see how the parser displays it!
response = client.agents.messages.create(
    agent_id=agent.id,
    input="请记住，我最喜欢的颜色是深蓝色，我是一个程序员。把这个写进你的核心记忆里。"
)

pretty_print_messages(response.messages)
```
